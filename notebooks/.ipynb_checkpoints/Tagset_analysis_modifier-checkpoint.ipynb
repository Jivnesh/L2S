### Variation: III

```python
case_list = ['nom','voc','acc','i','dat','abl','g','loc']
gender_list = ['n','f','m','*']
person_list = ['1','2','3']
no_list = ['du','sg','pl']
pops = [' ac',' ps']
indeclinable = ['ind','prep','interj','prep','conj','part','abs','ca abs']
```

```python
import re
store =[]
for ma in ma_set:
    temp = re.sub("([\(\[]).*?([\)\]])", "\g<1>\g<2>", ma).replace('[] ','').strip(' []')
    temp = temp.split('.')
    if temp[-1] == '':
        temp.pop(-1)
    # Remove active passive
    case='-'
    no='-'
    person='-'
    gender='-'
    tense='-'
    for a,b in enumerate(temp):
        if b in pops:
            temp.pop(a)
    # Get gender
    for a,b in enumerate(temp):
        if b.strip() in gender_list:
            gender = b.strip()
            temp.pop(a)
    # Get case
    for a,b in enumerate(temp):
        if b.strip() in case_list:
            case = b.strip()
            temp.pop(a)
    # Get person
    for a,b in enumerate(temp):
        if b.strip() in person_list:
            person = b.strip()
            temp.pop(a)
    # Get no
    for a,b in enumerate(temp):
        if b.strip() in no_list:
            no = b.strip()
            temp.pop(a)
    # Get Tense
    for b in temp:
        tense=tense+ ' '+b.strip()
    tense=tense.strip()
    
    store.append([case,gender,tense,no,person])
```

### Variation: IV

```python
case_list = ['nom','voc','acc','i','inst','dat','abl','g','loc']
gender_list = ['n','f','m','*']
person_list = ['1','2','3']
no_list = ['du','sg','pl']
pops = [' ac',' ps']
```

```python
import re
store =[]
coarse_to_ma_dict =dict()
for ma in ma_set:
    temp = re.sub("([\(\[]).*?([\)\]])", "\g<1>\g<2>", ma).replace('[] ','').strip(' []')
    temp = temp.split('.')
    if temp[-1] == '':
        temp.pop(-1)
    # Remove active passive
    case='-'
    no='-'
    person='-'
    gender='-'
    tense='-'
    coarse='-'
    for a,b in enumerate(temp):
        if b in pops:
            temp.pop(a)
    # Get gender
    for a,b in enumerate(temp):
        if b.strip() in gender_list:
            gender = b.strip()
            temp.pop(a)
    # Get case
    for a,b in enumerate(temp):
        if b.strip() in case_list:
            case = b.strip()
            temp.pop(a)
    if case!= '-':
        coarse ='Noun'
    # Get person
    for a,b in enumerate(temp):
        if b.strip() in person_list:
            person = b.strip()
            temp.pop(a)
    # Get no
    for a,b in enumerate(temp):
        if b.strip() in no_list:
            no = b.strip()
            temp.pop(a)
    # Get Tense
    flag = 0
    for b in temp:
        if flag == 0:
            tense=''
            flag=1
        tense=tense+ ' '+b.strip()
    tense=tense.strip()
    if tense == 'adv':
        coarse = 'adv'
    for ind in indeclinable:
        if tense == ind:
            coarse = 'Ind'
    if tense!='-' and coarse=='-':
        if person !='-' and no!='-':
            coarse = 'FV'
        else
            coarse = 'IV'
    if coarse not in coarse_to_ma_dict.keys():
        coarse_to_ma_dict[coarse] = []
    else:
        coarse_to_ma_dict[coarse].append(ma)
    store.append([case,gender,tense,no,person,coarse])
```
